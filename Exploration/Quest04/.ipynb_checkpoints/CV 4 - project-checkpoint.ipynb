```python
import cv2
import numpy as np
import torch
import torchvision.transforms as T
from torchvision.models.segmentation import deeplabv3_resnet101
import matplotlib.pyplot as plt
import os

print(cv2.__version__)
```

# 1. 인물 모드 문제점 찾기
## Step 1. 인물모드 직접 해보기

```python
import os
import cv2
import matplotlib.pyplot as plt

home = os.getenv('HOME')
base = f"{home}/work/project/Exploration/Quest04/human_segmentation/images"

# 이미지 경로
paths = [
    f"{base}/my_image.jpg",          # SULLYOON
    f"{base}/IMG.jpg",               # cat
    f"{base}/background_sokcho.jpg"  # background
]
titles = ["SULLYOON", "cat", "background"]

# 이미지 불러오기 및 RGB 변환
imgs = [cv2.cvtColor(cv2.imread(p), cv2.COLOR_BGR2RGB) for p in paths]

for i in range(3):
    print(f"{titles[i]} 이미지 크기: {imgs[i].shape}")

# 시각화
plt.figure(figsize=(15, 5))
for i in range(3):
    plt.subplot(1, 3, i + 1)
    plt.imshow(imgs[i])
    plt.title(f'{titles[i]}')
    
plt.show()
```

```python
# 모델 불러오기
model = deeplabv3_resnet101(pretrained=True).eval()

# 변환 정의
transform = T.Compose([
    T.ToPILImage(),
    T.Resize((520, 520)),
    T.ToTensor(),
])

# 1번째, 2번째 이미지 선택
selected_imgs = imgs[:2]

output_masks = []

for i, img in enumerate(selected_imgs):
    input_tensor = transform(img).unsqueeze(0)
    
    with torch.no_grad():
        output = model(input_tensor)["out"][0]
        mask = output.argmax(0).byte().cpu().numpy()
    
    output_masks.append(mask)
    print(f"이미지 {i+1} 추론 마스크 크기: {mask.shape}")
```

```python
# 추론 마스크(output_masks) 2장에 대해 처리
for i, mask in enumerate(output_masks):
    # 원본 이미지 크기 가져오기
    orig_img = imgs[i]  # imgs[0], imgs[1] 사용
    h, w = orig_img.shape[:2]

    # 마스크를 원본 크기로 Resize
    mask_resized = cv2.resize(mask, (w, h), interpolation=cv2.INTER_NEAREST)
    print(f"이미지 {i+1} 추론 마스크 크기 (After Resize): {mask_resized.shape}")

    # 고유 클래스 확인
    unique_classes = np.unique(mask_resized)
    print(f"이미지 {i+1} 예측된 클래스 ID: {unique_classes}")

    # 마지막 클래스 선택 (대상 클래스)
    target_class_id = unique_classes[-1]

    # 마스크 생성
    seg_map = (mask_resized == target_class_id)
    img_mask = seg_map.astype(np.uint8) * 255

    # 시각화
    plt.figure(figsize=(6, 6))
    plt.title(f"image {i+1} mask")
    plt.imshow(img_mask, cmap='gray')
    plt.axis('off')
    plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 배경 이미지를 두 이미지 크기에 맞춰 Resize
background_resized_list = [
    cv2.resize(background_img, (imgs[i].shape[1], imgs[i].shape[0]))
    for i in range(2)
]

result_imgs = []

for i in range(2):
    orig_img = imgs[i]
    mask = output_masks[i]

    # 마스크를 원본 이미지 크기로 Resize
    mask_resized = cv2.resize(mask, (orig_img.shape[1], orig_img.shape[0]), interpolation=cv2.INTER_NEAREST)

    # 마지막 클래스 선택 후 마스크 생성
    target_class_id = np.unique(mask_resized)[-1]
    seg_map = (mask_resized == target_class_id)
    img_mask = seg_map.astype(np.uint8) * 255

    # 마스크 기반 합성
    img_mask_color = cv2.cvtColor(img_mask, cv2.COLOR_GRAY2BGR)
    result_img = np.where(img_mask_color == 255, orig_img, background_resized_list[i])

    result_imgs.append(result_img)


plt.figure(figsize=(12, 6))
for i, img in enumerate(result_imgs):
    plt.subplot(1, 2, i + 1)
    plt.imshow(img)
    plt.title(f"image {i+1} synthetic results")
plt.show()
```

## Step 2. 사진에서 문제점 찾기

- 위 두 사진을 보면 어색한 부분이 보인다.
- 사람 이미지에서는 오른쪽 팔 부분이 잘려있고, 왼쪽 팔 부분은 그 양 옆 빈 여백부분을 사람 mask로 적용되서 진행이 되었다.
- 고양이는 번쩍 든 팔과 얼굴 사이 빈 여백을 고양이로 취급하여 mask가 적용되었다.
- 무엇보다 전체적으로 mask 처리한 라인이 다소 어색하다는 것을 볼 수 있다.
