```python
pokemon = pokemon.replace(to_replace="Nidoran♀", value="Nidoran X")
pokemon = pokemon.replace(to_replace="Nidoran♂", value="Nidoran Y")
pokemon = pokemon.replace(to_replace="Farfetch'd", value="Farfetchd")
pokemon = pokemon.replace(to_replace="Mr. Mime", value="Mr Mime")
pokemon = pokemon.replace(to_replace="Porygon2", value="Porygon Two")
pokemon = pokemon.replace(to_replace="Ho-oh", value="Ho Oh")
pokemon = pokemon.replace(to_replace="Mime Jr.", value="Mime Jr")
pokemon = pokemon.replace(to_replace="Porygon-Z", value="Porygon Z")
pokemon = pokemon.replace(to_replace="Zygarde50% Forme", value="Zygarde Forme")

pokemon.loc[list(pokemon[pokemon["name_isalpha"] == False].index)]
```

```python
# Q. 바꿔준 'Name' 컬럼으로 'Name_nospace'를 만들고, 다시 isalpha()로 체크해봅시다.
pokemon["Name_nospace"] = pokemon["Name"].apply(lambda i: i.replace(" ", ""))
pokemon["name_isalpha"] = pokemon["Name_nospace"].apply(lambda i: i.isalpha())

pokemon[pokemon["name_isalpha"] == False]
```

```python
import re

# 
def tokenize(name):
    name_split = name.split(" ")
    tokens = []
    for part_name in name_split:
        a = re.findall('[A-Z][a-z]*', part_name)
        tokens.extend(a)

    return np.array(tokens)


all_tokens = list(pokemon[pokemon["Legendary"] == True]["Name"].apply(tokenize).values)

token_set = []
for token in all_tokens:
    token_set.extend(token)

print(len(set(token_set)))
print(token_set)
```

```python
from collections import Counter

most_common = Counter(token_set).most_common(10) # 가장 많은 요소와 등장 횟수가 무엇인지를 정렬
most_common
```

```python
for token, _ in most_common:
    pokemon[f"{token}"] = pokemon["Name"].str.contains(token)

pokemon.head(10)
```

```python
# Type1 과 2 동시에 one-hot encoding
types12 = list(set(list(pokemon["Type 1"].unique()) + list(pokemon["Type 2"].unique())))
for t in types12:
    pokemon[t] = (pokemon["Type 1"] == t) | (pokemon["Type 2"] == t)

pokemon[[["Type 1", "Type 2"] + types12][0]].head()
```

```python
features = ['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation']
target = 'Legendary'

X = df[features]
print(X.shape)
print(X.head())

y = df[target]
print(y.shape)
print(y.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
%%time
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=25)

model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve, auc

fpr, tpr, thresholds = roc_curve(y_test, y_pred)
# thresholds는 잘 안 쓰이기에, 그냥 _로 표시하고 지나가도 됨

roc_auc = auc(fpr, tpr)  # 영역의 넓이 (0~1 출력)
```

```python
print(f'auc : {roc_auc}')
plt.figure(figsize=(8, 8)) # 그래프 크기 설정

# ROC 커브 그리기
# label에 AUC 점수를 소수점 2자리까지 표시
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')

# 랜덤 분류기(성능이 없는 모델)의 ROC 커브인 y=x 직선 그리기
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')

# 그래프 설정
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('Original data train (ROC) Curve', fontsize=15)
plt.legend(loc="lower right") # 범례 표시
plt.grid(True)
plt.show()
```
